```python
def crawling():
    #현재위치
    loc = os.getcwd()
    
    #크롬 드라이버 불러오기
    driver = webdriver.Chrome("{}/chromedriver.exe".format(loc))

    # 네이버 뉴스 URL
    url = 'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=100'

    # 드라이버 실행
    driver.get(url)

    # output list
    news_list = []

    # 7개 카테고리
    for i in range(1,7):
        # 카테고리 클릭
        driver.find_elements(By.CLASS_NAME,'tx')[i].click()
        time.sleep(1)
        # 카테고리 이름 추출
        category = driver.find_elements(By.CLASS_NAME,'tx')[i].text
        # 5개 뉴스 추출( 5개씩 CSS_SELECTOR 형태가 바뀌기 때문에 일단 5개만 )
        for j in range(1,6):
            # 뉴스 제목 추출
            title = driver.find_elements(By.CSS_SELECTOR,'#section_body > ul.type06_headline > li:nth-child({}) > dl > dt:nth-child(2) > a'.format(j))[0].text
            
            # 뉴스 클릭
            driver.find_elements(By.CSS_SELECTOR,'#section_body > ul.type06_headline > li:nth-child({}) > dl > dt:nth-child(2) > a'.format(j))[0].click()
            time.sleep(1)
            
            news_url = driver.current_url
            
            #뉴스 내용 추출
            content = driver.find_elements(By.CSS_SELECTOR, '#dic_area')[0].text
            
            # 명사추출 모델
            okt = Okt()
            noun = okt.nouns(content)
            # 명사 중 count가 많은 20개만 일단 추출
            noun_list = Counter(noun).most_common(20)
            # 불용어 제거
            noun_pre_list = [i for i in noun_list if len(i[0]) > 1 ]
            
            # 리스트에 삽입
            news_list.append([category,title,noun_pre_list, news_url])
            
            # 페이지 뒤로가기
            driver.back()
            time.sleep(1)
            
    return news_list
```
